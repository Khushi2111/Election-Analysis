```python
plt.figure(figsize=(15,8))
sns.set(style="darkgrid")
L3=list(politics["EDUCATION"].value_counts().keys())
L4=list(politics["EDUCATION"].value_counts())
sns.barplot(data=politics,x=L4,y=L3,palette="ch:s=.25,rot=-.25")
plt.xlabel("count")
plt.ylabel("Education")
plt.title("Education Level of the Candidates")
```
